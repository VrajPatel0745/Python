```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split

data={
    'StudyHours':[1,2,3,4,5],
    'TestScores':[40,50,60,70,80]
}
#Standard Scaling
df=pd.DataFrame(data)
standard_scaler=StandardScaler()
scaled_data=standard_scaler.fit_transform(df)
# z=(x-mean)/std_dev  [-x, +x]
print("Standard Scaled Data")
print(pd.DataFrame(scaled_data,columns=['StudyHours','TestScores']))

#Min-Max Scaling
min_max_scaler=MinMaxScaler()
scaled_data=min_max_scaler.fit_transform(df)
# x_scaled=(x - min)/(max - min)   0-1
print("Min-Max Scaled Data(0-1)")
print(pd.DataFrame(scaled_data,columns=['StudyHours','TestScores']))
```

### Spliting of Training and Testing data

```python
X=df[['StudyHours']]
Y=df[['TestScores']]
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

print("X_train:\n",X_train)
print("X_test:\n",X_test)
print("y_train:\n",y_train)
print("y_test:\n",y_test)
```
