```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

### Label Encodeing

```python
df=pd.read_csv('data.csv')
df_label=df.copy()
le=LabelEncoder()
df_label["Gender_Encoded"]=le.fit_transform(df_label["Gender"])
df_label["Passed_Encoded"]=le.fit_transform(df_label["Passed"])
print("\nLabel Encoded Data")
print(df_label[['Name','Gender','Gender_Encoded','Passed','Passed_Encoded']].head())
```

### One Hot Encoding
#### .get_dummies - get_dummies() is a pandas function used to convert categorical (text) columns into numerical columns using One-Hot Encoding.

#### Machine Learning models cannot work with text, so we convert categories into 0 / 1.

```python
df_encoded=pd.get_dummies(df_label,columns=['City'])
print("\nOne Hot Encoded Data")
print(df_encoded.head())
```

### Pandas .astype(newDataType) astype is a pandas method used to change (convert) the data type of a column or entire DataFrame.

```python
df_encoded = pd.get_dummies(df_label, columns=['City'])

dummy_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[dummy_cols] = df_encoded[dummy_cols].astype(int)

print(df_encoded)
print(df_label)
```
