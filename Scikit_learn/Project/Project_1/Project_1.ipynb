```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
```

```python
data=pd.read_csv("student.csv")
x=data[['Hours']] #double brackets =2D input
y=data['Score'] # Target colums

model=LinearRegression()
model.fit(x,y)
predicated_score=model.predict(x)
print(predicated_score)
mae=mean_absolute_error(y,predicated_score)
mse=mean_squared_error(y,predicated_score)
rmsr=np.sqrt(mse)

print("MAE on average off by:",mae)
print("MSE on average off by:",mse)
print("RMSR on average off by:",rmsr)

new_hour=float(input("Enter a hour="))
new_pred=model.predict([[7]])
print(f"Predication for{new_hour} is score={new_pred}")
```
