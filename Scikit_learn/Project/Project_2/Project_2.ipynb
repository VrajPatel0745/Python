```python
import pandas as pd
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('student_data.csv')
x=data[["studytime"]]
y=data["G1"]

model=LinearRegression()
model.fit(x,y)
predicated_score=model.predict(x)
mae=mean_absolute_error(y,predicated_score)
mse=mean_squared_error(y,predicated_score)
rmsr=np.sqrt(mse)
r2_score=r2_score(y,predicated_score)

print("MAE on average off by:",mae)
print("MSE on average off by:",mse)
print("RMSR on average off by:",rmsr)
print("R2_score on average off by:",r2_score)
```

```python
plt.figure(figsize=(10,6))
plt.hist(data["studytime"],bins=30,color='skyblue',edgecolor='black')
plt.title("Distribution of the Final Exam Scores")
plt.xlabel("Final Exam Score")
plt.ylabel("Numbers of students")
plt.grid(True)
plt.show()

# Scatter + Regression Line
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='skyblue',label='Actual_Scores')
plt.plot(x,predicated_score,color='red',label="Predicted Scores (Regression Line)")
plt.title("Model Prediction VS Actual Score")
plt.xlabel("Study Hours per Week")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()

new_hours=9
predicated_new_score=model.predict([[new_hours]])
print(f"Predicted Final Score for {new_hours} Hours is {predicated_new_score}")
# Scatter + Regression Line
```
