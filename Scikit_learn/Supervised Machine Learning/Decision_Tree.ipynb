```python
from sklearn.tree import DecisionTreeClassifier
# Decision Tree Classification
```

```python
x=[
  [7,2], #Apple
  [8,3], #Apple
  [9,8], #Orange
  [10,9] #Orange
]
y=['Apple','Apple','Orange','Orange']
model=DecisionTreeClassifier()
model.fit(x,y)
size=float(input("Enter size of fruit in cm: "))
weight=float(input("Enter color shades of fruit (1-10): "))
fruit=model.predict([[size,weight]])[0]
if fruit=='Apple':
    print("The fruit is Apple")
else:
    print("The fruit is Orange")
```
