## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Linear Regression 
# 1-finds pattern in old data
# 2-Straight line equation
# 3-y=mx+c
# [[values]] 2D array
# its not about the line, its about the story
# 2- 5 rows use start
# 3- accuracy is not goal
model=LogisticRegression()
x=[[1],[2],[3],[4],[5]]
y=[0,0,0,1,1 ]
model.fit(x,y)
hours=float(input("Enter hours studied: "))
predicted_marks=model.predict([[hours]])[0]
if predicted_marks==1: 
    print("Based on hour you are likely to Pass") 
else:
    print("Based on hour you are likely to Fail")

```

## KNN Classification


```python
from sklearn.neighbors import KNeighborsClassifier
# KNN Classification

```

```python

x=[
    [100,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]
#0 =Apple , 1= Orange
y=[0,0,0,1,1,1]
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
weight=float(input("Enter weight of fruit(grams): "))
size=float(input("Enter size of fruit(cm): "))
predicted_marks=model.predict([[weight,size]])[0]
if predicted_marks==1: 
    print("Based on weight and size it is an Orange")
else:
    print("Based on weight and size it is an Apple")
```
