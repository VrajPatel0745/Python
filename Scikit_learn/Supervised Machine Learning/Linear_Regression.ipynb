```python
from sklearn.linear_model import LinearRegression
```

```python
# Linear Regression 
# 1-finds pattern in old data
# 2-Straight line equation
# 3-y=mx+c
# [[values]] 2D array
# its not about the line, its about the story
# 2- 5 rows use start
# 3- accuracy is not goal
model=LinearRegression()
x=[[1],[2],[3],[4],[5]]
y=[40,50,65,75,80]
model.fit(x,y)
hours=float(input("Enter hours studied: "))
predicted_marks=model.predict([[hours]])
print(f"Predicted marks for studying {hours} hours: {predicted_marks}")

```
