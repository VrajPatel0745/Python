### Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
# MAE
# find absolute differnce 
# add total
# divide by total number of items

#MSE
# find absolute difference
# find sqaure
# add total of sqaure of absolute differnce
#  divide by total number of items

# RMSE
# square root of MSE
  
real_scores=[90,60,80,100]
predicated_scores=[85,70,70,95]

mae=mean_absolute_error(real_scores,predicated_scores)
mse=mean_squared_error(real_scores,predicated_scores)
rmsr=np.sqrt(mse)

print("MAE on average off by:",mae)
print("MSE on average off by:",mse)
print("RMSR on average off by:",rmsr)
```
