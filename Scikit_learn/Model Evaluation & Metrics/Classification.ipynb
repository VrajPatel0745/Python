```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score
```

```python
# True appears( What acually happened)
X_true=[1,0,1,1,0,1,0]
# Predicted appears ( What model predicted)
X_pred=[1,0,1,0,0,1,1]

#evaluation

print("Accuracy:",accuracy_score(X_true,X_pred))
print("Precision:",precision_score(X_true,X_pred))
print("Recall:",recall_score(X_true,X_pred))
print("F1 Score:",f1_score(X_true,X_pred))
```
