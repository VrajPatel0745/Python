## Advance operation with Business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id,2021,2022,2023,2024]
sales_data = np.array([
    [1,150000,180000,220000,250000],  #Paradise Biryani
    [2,120000,140000,160000,190000],  #Beijing Bites
    [3,200000,230000,260000,300000],  # Pizza Hub
    [4,180000,210000,240000,270000],  #Burger Point
    [5,160000,185000,205000,230000]   #Chai Point
])

print("===Zomato Restaurant Sales Data Analysis===")
print("\n Sales data shape:",sales_data.shape)
print("Sample data for 1st 3 restaurants:\n",sales_data[0:3])
print("\n",sales_data[:,1:])
```

```python
# total sales for each year
print(np.sum(sales_data,axis=0))  # total sales across all years and restaurants
year_total=np.sum(sales_data[:,1:],axis=1)
print("Total sales for each restaurant across years:\n",year_total)
# for all restaurants in all years
yearly_sales=np.sum(sales_data[:,1:],axis=0)
print("Yearly sales across all restaurants:\n",yearly_sales)
# minimum sales per restaurant
min_sales=np.min(sales_data[:,1:],axis=1)
print("Minimum sales per restaurant across years:\n",min_sales)
# maximum sales per restaurant
max_sales=np.max(sales_data[:,1:],axis=0)
print("Maximum sales per year across all restaurants:\n",max_sales)
# average sales per restaurant
avg_sales=np.mean(sales_data[:,1:],axis=1)
print("Average sales per restaurant across years:\n",avg_sales)
# standard deviation of sales per restaurant
std_sales=np.std(sales_data[:,1:],axis=1)
print("Standard deviation of sales per restaurant across years:\n",std_sales)

```

```python
cumulative_sales=np.cumsum(sales_data[:,1:],axis=1)
print("Cumulative sales per restaurant across years:\n",cumulative_sales)
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumulative_sales,axis=0))
plt.title("Average Cumulative Sales Across Restaurants Over Years")
plt.xlabel("Years (2021-2024)")
plt.ylabel("Average Cumulative Sales")
plt.grid()
plt.show()
```

```python
vector1=np.array([1,2,3,4,5])
vector2=np.array([6,7,8,9,10])

print("Addition:",vector1+vector2)
print("\nMultiplication:",vector1*vector2)
print("\nDot Product:",np.dot(vector1,vector2))
angle=np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("\nAngle between vectors (in radians):",angle)
print("\nAngle between vectors (in degrees):",np.degrees(angle))
```

### Vectorized Operation

```python
restaurant_types=np.array(['Biryani','Chinese','Italian','Fast Food','Beverages'])
print("Restaurant types:\n",restaurant_types)
vectorized_upper=np.vectorize(str.upper)
print("Restaurant types in uppercase:\n",vectorized_upper(restaurant_types))
```
