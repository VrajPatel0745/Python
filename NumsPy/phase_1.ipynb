## Numpy array and basics


```python
import numpy as np
```

Creating array from list

```python
arr_1d = np.array([1, 2, 3])
print("1D array:",arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D array:\n",arr_2d)
```

### List vs numpy array

```python
py_list=[1,2,3]
print("Python List Multiplication ",py_list*2) 
np_array = np.array([1,2,3]) # element-wise operations
print("Numpy Array Multiplication ",np_array*2)

import time
start = time.time()
py_list = list(i*2 for i in range(1000000))
print ("Python List Time: ", time.time() - start)
start = time.time()
np_array = np.arange(1000000)*2
print ("Numpy Array Time: ", time.time() - start)
```

### creating array from scratch

```python
zeros=np.zeros((3,4))
print("Zeros Array:\n",zeros)

ones=np.ones((3,4))
print("Ones Array:\n",ones)

full=np.full((2,2),7)
print("Full Array:\n",full)

random = np.random.random((2,3))
print("Random Array:\n",random)

sequence = np.arange(0,10,2)
print("Sequence Array:",sequence)

```

### Vector, Matrix and Tensor

```python
vector=np.array([1,2,3])
print("Vector:",vector)

matrix=np.array([[1,2,3],[4,5,6]])
print("Matrix:\n",matrix)

tensor=np.array([[[1,2],[3,4]],
                 [[5,6],[7,8]]])
print("Tensor:\n",tensor)
```

### Array properties


```python
arr=np.array([[1,2,3],
              [4,5,6]])
print("Shape",arr.shape)
print("Dimensions",arr.ndim)
print("Size",arr.size)
print("Data Type",arr.dtype)
#Float data Types
arr=np.array([[1,2,3],
              [4,5,6.8]])
print("Shape",arr.shape)
print("Dimensions",arr.ndim)
print("Size",arr.size)
print("Data Type",arr.dtype)
```

### Array Reshaping

```python
arr=np.arange(12)
print("Original Array:",arr)

reshaped=arr.reshape((3,4))
print("\n Reshaped Array (3x4):\n",reshaped)

# flatten (returns copy)
flattened=reshaped.flatten()
print("\n Flattened Array:",flattened)

# ravel (returns view , instead of copy)
raveled=reshaped.ravel()
print("\n Raveled Array:",raveled)

#Transpose
transpose =reshaped.T
print("\n Transposed Array:\n",transpose)
```
